#Part 4: Marketing Strategy

```python
daily_trends = df.groupby('Order Date').agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Discount': 'mean'
}).reset_index()
daily_trends
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.scatterplot(x='Discount', y='Sales', data=daily_trends)
plt.title('Daily Discount vs. Total Sales')
plt.xlabel('Daily Discount')
plt.ylabel('Total Sales')

plt.subplot(1, 2, 2)
sns.scatterplot(x='Discount', y='Profit', data=daily_trends)
plt.title('Daily Discount vs. Total Profit')
plt.xlabel('Daily Discount')
plt.ylabel('Total Profit')

plt.tight_layout()
plt.show()
```

```python
discount_sales_corr = daily_trends['Discount'].corr(daily_trends['Sales'])
discount_profit_corr = daily_trends['Discount'].corr(daily_trends['Profit'])

print(f'Pearson correlation between Daily Discount and Total Sales: {discount_sales_corr:.4f}')
print(f'Pearson correlation between Daily Discount and Total Profit: {discount_profit_corr:.4f}')
```

*   The Pearson correlation between daily discount and total sales is -0.0447,  a negligible linear relationship.
*   The Pearson correlation between daily discount and total profit is -0.2284, a weak negative linear relationship.

*   Discounts dont make much off a difference on sales. Although sales would be expected to increase during a disocunt, if anything we see a negligible decrease in sales. The decrease in profit although more noticable is not enough for a strong linear relationship. Discounts need to be done right to not lose too much of profit margin while still increasing sales.

*   Discounts should be used on inventory that is not selling. It is better to make some money than none at all. Looking closely at the data, there is a slight increase in profit at the 10 - 20% range, compared to any other range. Disconts on the sub categories in the bottom 5 that are still profitable can be good places to start. The results of those discounts can answer if the category itself isn't disearbale or the price that they are selling at. Even discounts on phones may be concered as although the sales numbers are large so are the profits. A small short time discount can bring customers through the door. Once they are in the store they may consider buying something else on discount.
